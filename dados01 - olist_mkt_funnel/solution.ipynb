```python
import pandas as pd
import numpy as np
```

Processamento de dados:

```python
# Tratando informações da tabela de todos os Leads 
df1 = pd.read_csv('olist_marketing_qualified_leads_dataset.csv', sep = ',')
```

```python
# Tratando informações da tabela de acordos firmados 
df2 = pd.read_csv('olist_closed_deals_dataset.csv', sep = ',')
```

```python
# Criando a coluna 'is_closed' na tabela de acodos fechados/firmados
df2['is_closed'] = True
```

- Criar tabela com todos os MQLs, com todas as informações disponíveis para cada um.

```python
# Criando a tabela com todas as informações disponíveis
tabela_mqls = pd.merge(df1, df2, how='left', on='mql_id')
```

- Criar uma coluna com quais MQLs fecharam acordo e quais foram perdidos.

```python
# Tratando NaN na coluna de 'is_closed' (True = Acordo Fechado / False = Acordo Perdido)
tabela_mqls['is_closed'] = tabela_mqls['is_closed'].fillna(False)
tabela_mqls.head(5)
```

Perguntas a serem respondidas:
- Qual foi a taxa de conversão total?

```python
#Total de Leads
total_leads = tabela_mqls.is_closed.count()
total_leads
```

```python
#quantidade de linhas NÃO nulas - Acordos fechados
closed_deal = tabela_mqls.is_closed.sum()
closed_deal
```

```python
tx_conv_total = closed_deal / total_leads
print("A taxa de conversão total foi de:", '{0:.3%}'.format(tx_conv_total))
```

- Qual foi a taxa de conversão de cada origem?

```python
conv_origem = tabela_mqls.groupby('origin')['is_closed'].count().to_frame('total')
conv_origem['closed'] = tabela_mqls.groupby('origin')['is_closed'].sum()
conv_origem['taxa'] = conv_origem['closed']/conv_origem['total']
conv_origem
```

```python
#Outra forma de fazer o exemplo de cima =) thanks prof! 
def perc(x):
    return (x==1).astype(int).sum() / x.count()

tabela_mqls.groupby('origin').agg({'is_closed':
                                                ['sum', 
                                                'count', 
                                                perc]})
```

- Qual foi a taxa de conversão de cada página inicial?

```python
conv_landing_page = tabela_mqls.groupby('landing_page_id')['is_closed'].count().to_frame('total')
conv_landing_page['closed'] = tabela_mqls.groupby('landing_page_id')['is_closed'].sum()
conv_landing_page['taxa'] = conv_landing_page['closed']/conv_landing_page['total']

conv_landing_page
```

- Para cada SR, qual a receita média declarada dos leads?

```python
tabela_mqls.groupby('sr_id')['declared_monthly_revenue'].mean().sort_values(ascending=False).to_frame("receita_media")
```

- Para cada SR, quantos de cada business_type a pessoa converteu?

```python
tabela_mqls.pivot_table(columns='business_type', index='sr_id', values='is_closed',aggfunc='sum').fillna(0)
```

- Para cada SR, quantos de cada lead_behaviour_profile a pessoa converteu?

```python
tabela_mqls.pivot_table(columns='lead_behaviour_profile', index='sr_id', values='is_closed',aggfunc='sum').fillna(0)
```

- Para cada SR, quantos de cada lead_type a pessoa converteu?

```python
tabela_mqls.pivot_table(columns='lead_type', index='sr_id', values='is_closed',aggfunc='sum').fillna(0)
```
