```python
import pandas as pd
```

Processamento de dados:

```python
# Tratando informações da tabela de todos os Leads 
df1 = pd.read_csv('olist_marketing_qualified_leads_dataset.csv', sep = ',')
```

```python
# Tratando informações da tabela de acordos firmados 
df2 = pd.read_csv('olist_closed_deals_dataset.csv', sep = ',')
```

```python
# Criando a coluna 'is_closed' na tabela de acodos fechados/firmados
df2['is_closed'] = True
```

- Criar tabela com todos os MQLs, com todas as informações disponíveis para cada um.

```python
# Criando a tabela com todas as informações disponíveis
tabela_mqls = pd.merge(df1, df2, how='left', on='mql_id')
```

- Criar uma coluna com quais MQLs fecharam acordo e quais foram perdidos.

```python
# Tratando NaN na coluna de 'is_closed' (True = Acordo Fechado / False = Acordo Perdido)
tabela_mqls['is_closed'] = tabela_mqls['is_closed'].fillna(False)
tabela_mqls.head(5)
```

Perguntas a serem respondidas:
- Qual foi a taxa de conversão total?

```python
#Total de Leads
total_leads = tabela_mqls.is_closed.count()
total_leads
```

```python
#quantidade de linhas NÃO nulas - Acordos fechados
closed_deal = tabela_mqls.is_closed.sum()
closed_deal
```

```python
tx_conv_total = closed_deal / total_leads
print("A taxa de conversão total foi de:", '{0:.3%}'.format(tx_conv_total))
```
